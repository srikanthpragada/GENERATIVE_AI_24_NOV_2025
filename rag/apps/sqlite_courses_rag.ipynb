```python
import sqlite3
from langchain_ollama import OllamaLLM, OllamaEmbeddings  # Import the OllamaLLM class
from langchain_community.vectorstores import FAISS
from tqdm import tqdm
```

```python
db_path = r"c:\classroom\genai\sqlite\courses.db"
```

```python
# Get data from database using DB API 2.0 
def get_courses():
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()
    cursor.execute("SELECT * FROM courses")
    courses = cursor.fetchall()
    conn.close()
    return courses
```

```python
def concate_course(course):
    course_id, title, description, duration, fee, prerequisite = course
    return f"Title: {title}\nDescription: {description}\nFee: {fee}\nDuration: {duration}\nPrerequisite: {prerequisite}"
```

```python
# Load courses data
courses = get_courses()

courses_list = []
for course in courses:
    courses_list.append(concate_course(course))
```

```python
print(courses_list)
```

```python
embeddings_model = OllamaEmbeddings(model="nomic-embed-text")
vectorstore = FAISS.from_texts(courses_list, embeddings_model)
```

```python
question = "Which course covers RAG"
matching_results = vectorstore.similarity_search(query, k=2)
```

```python
context = ""
for match in matching_results:
    context += match.page_content + "\n\n"
```

```python
prompt = f"""
You are a helpful assistant. You will be provided with a context and a question. Your task is to answer the question based on the context.
Context : 
{context}

Question: {question}
"""
```

```python
model = OllamaLLM(model="llama3.2:latest")
response = model.invoke(prompt)
print(response)
```
